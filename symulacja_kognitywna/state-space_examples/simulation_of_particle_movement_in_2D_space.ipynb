### Prosty program symulujący ruch cząsteczek w przestrzeni 2D

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parametry symulacji
time_simulation = 10.0 # czas trwania symulacji wyrazony w sekundach
time_step = 0.1 # krok czasowy wyrazony w sekundach
v_x = 1.0 # predkosc w kierunku X
v_y = 0.5 # predkosc w kierunku Y
```

```python
# Inicjalizacja polozenia i czasu 
x, y, t = 0.0, 0.0, 0.0
```

```python
# Listy do przechowywania śledzonych pozycji
x_possition, y_possition = [], []
```

```python
while t <= time_simulation:
    x_possition.append(x)
    y_possition.append(y)
    
    # Aktualizacja położenia
    x += v_x * time_step
    y += v_y * time_step

    t += time_step

# Wykres przedstawiajacy trajektorie ruchu
plt.plot(x_possition, y_possition)
plt.xlabel("X"), plt.ylabel("Y")
plt.title('Symulacja ruchu cząstki w przestrzeni 2D')
plt.grid(True)
plt.show()
```
